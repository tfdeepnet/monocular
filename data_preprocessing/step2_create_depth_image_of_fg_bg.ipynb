### Compress Dataset folder to zip format

```
import os
import zipfile
def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))
zipf = zipfile.ZipFile('/content/DenseDepth/Dataset.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('/content/DenseDepth/Dataset', zipf)
zipf.close()
```
